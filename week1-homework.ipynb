Q.6 Does the median value change after filling missing values?


```python
cylinders = df['Engine Cylinders']
median_before = cylinders.median()
mode = cylinders.mode()
cylinders.fillna(mode, inplace=False)
median_after = cylinders.median()
```

```python
if(median_before == median_after):
    print("Yes")
else:
    print("No")
```

Q.7 Value of the first element of w.

```python
lotus = df[df['Make'] == 'Lotus']
lotus = lotus[['Engine HP', 'Engine Cylinders']]
lotus = lotus.drop_duplicates()
lotus
```

```python
X = lotus.values
print(X)
```

```python
XTX = np.matmul(X.T, X)
XTX = XTX.astype('int')
XTX
```

```python
XTX_inv = np.linalg.inv(XTX)
XTX_inv
```

```python
y = np.array([1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800])
w = XTX_inv.dot(X.T).dot(y)

w[0]
```
